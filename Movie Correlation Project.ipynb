```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots


# Read in the data

df = pd.read_csv(r'C:\Users\navin\Downloads\movies.csv')

```

```python
# Let's look at the data

df.head()
```

```python
# Checking to see if data points are missing any information

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
# Taking care of missing data
df = df.dropna(how='any',axis=0) 
```

```python
df.head()
```

```python
# Data types for our columns
df.dtypes
```

```python
# Editing data types that have decimals (.0)
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df
```

```python
# Creating correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(str)

#df['release_date'] = df.apply(lambda x: x['released'][0:x['released'].find(' (')],axis=1)
#df['release_date'] = pd.to_datetime(df['release_date'], infer_datetime_format=True)
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```
